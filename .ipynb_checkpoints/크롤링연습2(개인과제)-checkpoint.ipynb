```python
url_list = []
driver = webdriver.Chrome('C:/selenium/chromedriver.exe')

for page in range(1, 78):

    url = "https://www.lawtimes.co.kr/Search/Law-Search.aspx?q=%EA%B3%A0%EB%A0%B9&page=" + str(page)
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    for i in range(1, 15):    
        elems = driver.find_elements_by_css_selector("#contents > div > section.sub-left > section > article:nth-child(" + str(i) + ") > a")
        link = [elem.get_attribute('href') for elem in elems]
        url_list.append(link)
        
```

```python
strings = url_list
```

```python
import itertools

url_list = list(itertools.chain(*strings))
```

```python
url_list
```

### 크롤링하기

```python
import requests
import re

content_list = []
title_list = []
author_list = []
date_list = []

for i in range(len(url_list)):
    try:
        res = requests.get(url_list[i])
        soup = BeautifulSoup(res.text, 'html.parser')
        content_result = str(soup.find('article', class_ = 'contents-txt'))
        content_result = re.sub('(<([^>]+)>)', '', content_result)
        content_list.append(content_result)
    
        title_result = str(soup.find('h2'))
        title_result = re.sub('(<([^>]+)>)', '', title_result)
        title_list.append(title_result)
    
        author_result = str(soup.find('a', class_ = 'author'))
        author_result = re.sub('(<([^>]+)>)', '', author_result)
        author_list.append(author_result)
    
        date_result = str(soup.find('time'))
        date_result = re.sub('(<([^>]+)>)', '', date_result)
        date_list.append(date_result)
    
        print(i)
        time.sleep(0.1)
    except:
        print('timeout error')
```

```python
#contents > header > h2

#contents > header > a.author

#contents > header > time

raw_result = {'title': title_list, 'author': author_list, 'content': content_list, 'date': date_list}
result = pd.DataFrame(raw_result)
```
